```python
import cv2
img = cv2.imread('a.jpg', 1)
h, w, c = img.shape
print('h : ', h, 'w : ', w, 'c : ', c)
p = img[100, 100] # 100,100 의 픽셀 값 추출
p  #[b, g, r]
```

```python
b_c = img[100, 100, 0]  #blue 색상값. 0:blue, 1:green, 2:red
b_c
```

```python
img[100, 100] = [255,255,255] #(100,100)위치의 픽셀을 흰색으로 변경
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img[:,:,2] = 255 # 모든 줄, 칸의 2번째 속성 값을 255로 변경
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
roi = img[100:200, 100:300]
cv2.imshow('roi', roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img[0:100, 200:400] = roi
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b,g,r = cv2.split(img) #비용 비쌈
img = cv2.merge((r,g,b))
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_blue = img[:,:,0]
img_blue
```

```python
from matplotlib import pyplot as plt 

img_blue = img[:,:,0] # 파랑만 추출
img_green = img[:,:,1] # 녹색만 추출
img_red = img[:,:,2] # 빨강만 추출
img = cv2.merge((img_red, img_green, img_blue)) # 순서 재배치
plt.imshow(img)
plt.show()
img_blue
```

```python
import cv2
src = cv2.imread('a.jpg')
img1 = src   #얕은 복사
img2 = src.copy()   #깊은 복사
img1[100:200, 100:200, 2] = 255
cv2.imshow('src', src)
cv2.imshow('img1', img1)
cv2.imshow('img2', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
src = cv2.imread('a.jpg')
img1 = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img2, cv2.COLOR_RGB2BGR)
cv2.imshow('src', src)
cv2.imshow('img1', img1)
cv2.imshow('img2', img2)
cv2.imshow('img3', img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
